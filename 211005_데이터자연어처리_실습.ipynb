```python
doc_ko = open("어바웃타임댓글데이터.csv").read()
doc_ko[1:1000]
```

```python
# OKT 클래스를 이용한 명사확인 Okt (11초 : 4286), Kkma (37초 : 5572)
from konlpy.tag import Okt   ### Okt
t = Okt()

#from konlpy.tag import Kkma  ### Okt
#t = Kkma()
doc_nouns = t.nouns(doc_ko)
print( len( doc_nouns ) ) #길이를 구한후 
doc_nouns #출력
```

```python
import nltk
```

```python
# nltk.Text()를 이용하여 nltk가 가지는 많은 기능을 사용 가능해짐.
ko = nltk.Text(doc_nouns, name="어바웃타임")
print(type(ko))
print(len(ko.tokens))
```

```python
### 단어들의 사용 횟수 확인 - 빈도 분석
type( ko.vocab() ), ko.vocab().most_common(20)
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
### 중복된 단어를 제거한 개수를 확인 #tokens
print(len(set(ko.tokens)))
```

```python
### 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['영화','진짜','그냥','정말','더','안','이','때','보고',
              '거','볼','것','수','점','임','그','최고','내',
              '정도','나','좀','말','함','걸','한국','뭐']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="어바웃타임")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
plt.figure(figsize=(15,8))
new_ko.dispersion_plot(['스토리', '중국', '양조위', '감동'])
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
abouttime_mask = np.array(Image.open("about_time.png"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
### 워드 클라우드 표현을 위한 데이터 생성
### 약간의 시간이 필요.
wc = WordCloud(background_color='white', 
               max_words=200,
               mask=abouttime_mask,   
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
