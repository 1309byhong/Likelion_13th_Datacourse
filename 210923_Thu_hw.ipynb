```python
dat = pd.read_csv("seoul_public_toilet.csv",encoding="EUCKR")
dat
```

```python
dat_tmp = dat.loc[:,['대명칭','WSG84Y좌표','WSG84X좌표']]
dat_tmp
```

```python
##03_일부데이터 획득하기
```

```python
###200개만 얻기
dat_200 = dat_tmp.iloc[0:200,:]
dat_200
```

```python
size = dat_200.name.count()
print(size) #크기
```

```python
print(dat_200.columns)
```

```python
import time
```

```python
##04_데이터 지도위에 표시하기
```

```python
# %%time
now = time.time()

m = folium.Map(location = [37.5447, 126.9828],zoom_start=13)

for i in range(size):
  folium.Marker(location = [dat_200.loc[i,'latitude'],
                            dat_200.loc[i,'longtitude']],
                            tooltip = dat_200.loc[i,'name'],
                icon = folium.Icon(color='orange',icon='info-sign')).add_to(m)

work_time = time.time() - now
m.save("seoul_data.html")
print(work_time)
m
```
